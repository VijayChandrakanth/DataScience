```python
df5 = pd.DataFrame([[25,'Black','SW',1],
                        [30,'Black','NW',1],
                        [45,'Red','NE',1],
                        [35,'Blond','SW',0],
                        [33,'Brown','SW',0],
                        [35,'Brown','NE',0],
                        [25,'Black','SW',0],
                        [52,'Brown','SE',0],
                        [21,'Blond','SW',1],
                        [24,'Blond','NE',0],
                        [48,'Black','SW',1],
                        [22,'Black','SW',0],
                        [27,'Brown','NE',0],
                        [28,'Black','SE',1],
                        [21,'Blond','SE',1],
                        [51,'Black','NW',0],
                        [23,'Brown','SE',1],                        
                       ]
                       ,columns =['AGE','HAIR COLOR','US REGION','VALIDATION'] )
```

```python
df6 = df5.loc[:7]
df7 = df5.loc[8:]
```

```python
df_Inner = df6.merge(df7, on='US REGION', how='inner')
df_Inner
```

```python
df_Left = df6.merge(df7, on='US REGION', how='left')
df_Left
```

```python
df_Right = df6.merge(df7, on='US REGION', how='right')
df_Right
```

```python
pd.concat([df6,df7], axis = 0)
```

```python
pd.concat([df6,df7], axis = 1)
```
